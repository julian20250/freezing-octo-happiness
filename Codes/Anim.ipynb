<h2>Period</h2>
<p>Knowing that the force is equal to $\frac{GMm}{r^2}$ and is equal to all the points of the orbit, we
calcule the aceleration via force: $a=\frac{GM}{r^2}$</p>
<p>Now we assume that the gravitational aceleration is equal to the centipete aceleration, so: $$\frac{GM}{r^2}=\frac{v^2}{r}$$ $$\frac{GM}{r}=v^2$$ $$v=wr \ \ and \ \ w=\frac{2\pi}{T}$$ $$\frac{GM}{r}=w^2r^2$$ $$\frac{GM}{r}=\frac{4\pi^2}{T^2} r^2$$ $$T^2=\frac{4\pi^2}{GM} r^3$$</p>
<p>Now assume that $T_e=400 \ period$, so, every planet has a period of $\frac{400 \times T_p}{T_e}$ in this code.</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.animation as animation
class SubplotAnimation(animation.TimedAnimation):
    def __init__(self):
        fig = plt.figure()
        ax1 = fig.add_subplot(1, 2, 1)
        ax2 = fig.add_subplot(2, 2, 2)
        ax3 = fig.add_subplot(2, 2, 4)

        self.t = np.linspace(0, 80, 400)
        self.x = np.cos(2 * np.pi * self.t / 10.)
        self.y = np.sin(2 * np.pi * self.t / 10.)
        self.z = 10 * self.t

        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        self.line1 = Line2D([], [], color='black')
        self.line1a = Line2D([], [], color='red', linewidth=2)
        self.line1e = Line2D([], [], color='red', marker='o', markeredgecolor='r')
        ax1.add_line(self.line1)
        ax1.add_line(self.line1a)
        ax1.add_line(self.line1e)
        ax1.set_xlim(-1, 1)
        ax1.set_ylim(-2, 2)
        ax1.set_aspect('equal', 'datalim')

        ax2.set_xlabel('y')
        ax2.set_ylabel('z')
        self.line2 = Line2D([], [], color='black')
        self.line2a = Line2D([], [], color='red', linewidth=2)
        self.line2e = Line2D([], [], color='red', marker='o', markeredgecolor='r')
        ax2.add_line(self.line2)
        ax2.add_line(self.line2a)
        ax2.add_line(self.line2e)
        ax2.set_xlim(-1, 1)
        ax2.set_ylim(0, 800)

        ax3.set_xlabel('x')
        ax3.set_ylabel('z')
        self.line3 = Line2D([], [], color='black')
        self.line3a = Line2D([], [], color='red', linewidth=2)
        self.line3e = Line2D([], [], color='red', marker='o', markeredgecolor='r')
        ax3.add_line(self.line3)
        ax3.add_line(self.line3a)
        ax3.add_line(self.line3e)
        ax3.set_xlim(-1, 1)
        ax3.set_ylim(0, 800)

        animation.TimedAnimation.__init__(self, fig, interval=50, blit=True)

    def _draw_frame(self, framedata):
        i = framedata
        head = i - 1
        head_len = 10
        head_slice = (self.t > self.t[i] - 1.0) & (self.t < self.t[i])

        self.line1.set_data(self.x[:i], self.y[:i])
        self.line1a.set_data(self.x[head_slice], self.y[head_slice])
        self.line1e.set_data(self.x[head], self.y[head])

        self.line2.set_data(self.y[:i], self.z[:i])
        self.line2a.set_data(self.y[head_slice], self.z[head_slice])
        self.line2e.set_data(self.y[head], self.z[head])

        self.line3.set_data(self.x[:i], self.z[:i])
        self.line3a.set_data(self.x[head_slice], self.z[head_slice])
        self.line3e.set_data(self.x[head], self.z[head])

        self._drawn_artists = [self.line1, self.line1a, self.line1e,
            self.line2, self.line2a, self.line2e,
            self.line3, self.line3a, self.line3e]

    def new_frame_seq(self):
        return iter(range(self.t.size))

    def _init_draw(self):
        lines =  [self.line1, self.line1a, self.line1e,
            self.line2, self.line2a, self.line2e,
            self.line3, self.line3a, self.line3e]
        for l in lines:
            l.set_data([], [])

ani = SubplotAnimation()
#ani.save('test_sub.mp4')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.animation as animation

G=6.6738e-11
M = 1.991e30
mercury= 5.79e10
venus=1.08e11
earth=1.496e11
Te= (((4*np.pi**2)/(G*M))*earth**3)**.5
ko= (Te**2/earth**3)
Tm= (ko*mercury**3)**.5
Tv= (ko*venus**3)**.5
Re =400
Rm= Tm*Re/Te
Rv= Tv*Re/Te
Dre=1
Drv=venus/earth
Drm=mercury/earth
```

```python


# This example uses subclassing, but there is no reason that the proper function
# couldn't be set up and then use FuncAnimation. The code is long, but not
# really complex. The length is due solely to the fact that there are a total
# of 9 lines that need to be changed for the animation as well as 3 subplots
# that need initial set up.
class SubplotAnimation(animation.TimedAnimation):
    def __init__(self):
        fig = plt.figure()
        ax1 = fig.add_subplot(1, 1, 1)


        self.t = np.linspace(0, 4000000, Re)
        self.p = np.linspace(0, 4000000, Rv)
        self.w = np.linspace(0, 4000000, Rm)
        self.x = Dre*np.cos(2 * np.pi * self.t / 10.)
        self.y = Dre*np.sin(2 * np.pi * self.t / 10.)
        self.x1 = Drv*np.cos(2 * np.pi * self.p / 10.)
        self.y1 = Drv*np.sin(2 * np.pi * self.p / 10.)
        self.y2= Drm*np.sin(2 * np.pi * self.w / 10.)
        self.x2= Drm*np.cos(2 * np.pi * self.w / 10.)
        

        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        
        self.line1 = Line2D([], [], color='black')
        self.line1e = Line2D([], [], color='green', marker='o', markeredgecolor='r')
        ax1.add_line(self.line1)
        ax1.add_line(self.line1e)
        
        self.line3= Line2D([], [], color='black')
        self.line3a = Line2D([], [], color='red', marker='o', markeredgecolor='r')
        ax1.add_line(self.line3)
        ax1.add_line(self.line3a)
        
        self.line2 = Line2D([], [], color='black')
        self.line2a = Line2D([], [], color='blue', marker='o', markeredgecolor='r')
        ax1.add_line(self.line2)
        ax1.add_line(self.line2a)
        ax1.set_xlim(-3, 3)
        ax1.set_ylim(-3, 3)
        ax1.set_aspect('equal', 'datalim')
        


        animation.TimedAnimation.__init__(self, fig, interval=100, blit=True)

    def _draw_frame(self, framedata):
        i = framedata
        head = i - 1

        self.line1.set_data(self.x[:i], self.y[:i])
        self.line1e.set_data(self.x[head], self.y[head])
        
        self.line2.set_data(self.x1[:i], self.y1[:i])
        self.line2a.set_data(self.x1[head], self.y1[head])
        
        self.line3.set_data(self.x2[:i], self.y2[:i])
        self.line3a.set_data(self.x2[head], self.y2[head])
        

        

        self._drawn_artists = [self.line1, self.line1e, self.line2, self.line2a,
                               self.line3, self.line3a]

    def new_frame_seq(self):
        return iter(range(self.t.size))

    def new_frame_seq1(self):
        return iter(range(self.p.size))
    
    def new_frame_seq1(self):
        return iter(range(self.w.size))
    
    def _init_draw(self):
        lines =  [self.line1, self.line1e, self.line2, self.line2a,
                  self.line3, self.line3a]
        for l in lines:
            l.set_data([], [])

ani = SubplotAnimation()
#ani.save('test_sub.mp4')
plt.show()
```
